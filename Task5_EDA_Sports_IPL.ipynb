The Indian Premier League (IPL) is a professional Twenty20 cricket league, contested by ten teams based out of ten Indian cities. The league was founded by the Board of Control for Cricket in India (BCCI) in 2007. It is usually held between March and May of every year and has an exclusive window in the ICC Future Tours Programme

### Analysis of Batsman

#### Most number of Runs by a Batsman

```python
mostruns = deliveries.groupby('batsman')['batsman_runs'].agg(sum).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
mostruns.head(10)
```

```python
labels = np.array(df_4s['batsman'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(df_4s['batsman_runs']), width=width)
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Batsman with most number of 4s.!",fontsize = 10)
autolabel(rects)
plt.show()
```

- __Dhawan__ is leading the chart by 526 4s, closely followed by __Raina__ and __Gambhir__

#### Most number of 6s by a Batsman

```python
labels = np.array(df_6s['batsman'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(df_6s['batsman_runs']), width=width)
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Batsman with most number of 6s.!",fontsize = 10)
autolabel(rects)
plt.show()
```

- __Gayle__ is way ahead of others, almost 113 Sixes more than __AB de Villiers__

### Analysis of Bowlers

#### Most number of Wickets by a Bowler
